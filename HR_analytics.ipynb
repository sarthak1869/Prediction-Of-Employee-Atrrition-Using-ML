```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('general_data.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['NumCompaniesWorked'].fillna(df['NumCompaniesWorked'].median(),inplace=True)
```

```python
df['TotalWorkingYears'].fillna(df['TotalWorkingYears'].median(),inplace=True)
```

```python
df['Attrition'].value_counts()
```

```python
df[df['Attrition']=='No']
```

```python
no_cat=df[df['Attrition']=='No'].sample(711)
yes_cat=df[df['Attrition']=='Yes']
```

```python
yes_cat
```

```python
df_resampled=pd.concat([no_cat,yes_cat],axis=0)
df_resampled
```

```python
cat=[]
num=[]
for i in df_resampled.columns:
    if df_resampled[i].dtype=='O':
        cat.append(i)
    else:
        num.append(i)
        
```

```python
cat
```

```python
num
```

```python
for i in cat:
    vc=df_resampled[i].value_counts()
    plt.bar(vc.index,height=vc,color='c')
    plt.xlabel(i)
    plt.xticks(rotation=90)
    plt.ylabel("Count")
    plt.show()
```

```python
df_resampled.drop('Over18',axis=1,inplace=True)
```

```python
 for i in num:
    fig, ax=plt.subplots(1,2)
    ax[0].hist(df_resampled[i],color='r')
    ax[1].boxplot(df_resampled[i])
    plt.suptitle(i)
    plt.show()
```

```python
for i in num:
    q1=df_resampled[i].quantile(0.25)
    q3=df_resampled[i].quantile(0.75)
    iqr=q3-q1
    lf=q1-1.5*iqr
    uf=q3+1.5*iqr
    df_resampled.loc[df_resampled[i]<lf,i]=lf
    df_resampled.loc[df_resampled[i]>uf,i]=uf
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df_resampled.reset_index(inplace=True)
```

```python
df_resampled
```

```python
df_resampled.columns
```

```python
ohe_dept=OneHotEncoder(sparse_output=False)
ohe_bt=OneHotEncoder(sparse_output=False)
ohe_ef=OneHotEncoder(sparse_output=False)
ohe_jr=OneHotEncoder(sparse_output=False)
ohe_ms=OneHotEncoder(sparse_output=False)
dept=ohe_dept.fit_transform(df_resampled[['Department']])
bt=ohe_bt.fit_transform(df_resampled[['BusinessTravel']])
ef=ohe_ef.fit_transform(df_resampled[['EducationField']])
jr=ohe_jr.fit_transform(df_resampled[['JobRole']])
ms=ohe_ms.fit_transform(df_resampled[['MaritalStatus']])
```

```python
dept_df=pd.DataFrame(dept,columns=ohe_dept.categories_)
bt_df=pd.DataFrame(bt,columns=ohe_bt.categories_)
ef_df=pd.DataFrame(ef,columns=ohe_ef.categories_)
jr_df=pd.DataFrame(jr,columns=ohe_jr.categories_)
ms_df=pd.DataFrame(ms,columns=ohe_ms.categories_)
```

```python
dept_df.drop('Human Resources',axis=1,inplace=True)
```

```python
dept_df
```

```python
bt_df.drop('Non-Travel',axis=1,inplace=True)
bt_df
```

```python
ef_df.drop('Human Resources',axis=1,inplace=True)
```

```python
ef_df
```

```python
jr_df.drop('Healthcare Representative',axis=1,inplace=True)
jr_df
```

```python
ms_df.drop('Divorced',axis=1,inplace=True)
ms_df
```

```python
df_resampled=pd.concat([df_resampled,dept_df,bt_df,ef_df,jr_df,ms_df],axis=1)
```

```python
df_resampled
```

```python
df_resampled['Attrition'].replace(to_replace=['Yes','No'],value=[1,0],inplace=True)
df_resampled['Gender'].replace(to_replace=['Male','Female'],value=[1,0],inplace=True)
```

```python
df_resampled
```

```python
X_df=df_resampled.drop(['BusinessTravel','Department','EducationField','Attrition','JobRole','MaritalStatus'],axis=1)
y=df_resampled['Attrition']
```

```python
X=X_df.values
Y=y.values
```

```python
X_df
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
y_train_pred=model.predict(X_train)
y_test_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,precision_score,recall_score,confusion_matrix
print("Model Performance: ")
print("Training accuracy: ",accuracy_score (y_train,y_train_pred))
print("Training precision: ", precision_score(y_train, y_train_pred))
print("Training recall: ", recall_score(y_train, y_train_pred))
print("Training confusion matrix: ",confusion_matrix(y_train, y_train_pred))
print("Training classification report: ",classification_report(y_train, y_train_pred))
```

```python
print("Model Performance: ")
print("Testing accuracy: ",accuracy_score (y_test,y_test_pred))
print("Testing precision: ", precision_score(y_test, y_test_pred))
print("Testing recall: ", recall_score(y_test, y_test_pred))
print("Testing confusion matrix: ",confusion_matrix(y_test, y_test_pred))
print("Testing classification report: ",classification_report(y_test, y_test_pred))
```
